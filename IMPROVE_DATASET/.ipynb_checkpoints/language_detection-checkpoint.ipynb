```python
#packages
import pandas as pd
from tqdm import tqdm, tqdm_notebook
tqdm_notebook().pandas()
from langdetect import detect
```

```python
def detect_language(string):
    try:
        language= detect(string)
    except:
        language= "unknown"
    return language
```

```python
df["language"]= df["quotation"].progress_apply(lambda x : detect_language(x))
```

WHOIS requests on the first URL of each quote

Importing needed packages

```python
import numpy as np
import pandas as pd
import bz2
import json
from whois import whois
from tqdm import tqdm
tqdm.pandas()
```

Function to perform WHOIS request

```python
def tryandget(url):
    try:
        who= whois(url)
        return [who.org, who.resgistrar, who.country, who.state, who.city] #[who.org, who.country, who.city]
    except:
        pass
    return [None, None, None, None, None]
```

Reading sample data

```python
df= pd.read_json("./../Sample.json.bz2",compression="bz2",lines=True)
```

Extracting domains from urls

```python
df["domain"]= df['urls'].apply(lambda x: "/".join(x[0].split('/',3)[:3]))
```

Creating a dataframe of unique domains

```python
df_domain= pd.DataFrame(df.value_counts("domain")).reset_index()
df_domain.describe()
```

Performing WHOIS requests on domains

```python
df_domain["origin"]= df_domain["domain"].apply(lambda x: tryandget(x))
```

Saving domain dataframe to pickle format

```python
df_domain.to_pickle("./whois_result.pkl")
df_domain.head(10)
```
